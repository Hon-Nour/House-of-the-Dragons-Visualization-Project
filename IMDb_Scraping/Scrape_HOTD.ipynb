```python
# Import necessary libraries
import pandas as pd
import requests
import openpyxl
from bs4 import BeautifulSoup
```

```python
# Get the data of the website
url = "https://www.imdb.com/title/tt11198330/episodes?ref_=tt_eps_sm"
response = requests.get(url)
hotd_episode_page = BeautifulSoup(response.content)

episodes = hotd_episode_page.findAll("div", attrs={"class": "info"})

# number_of_episodes = len(episodes)
# number_of_episodes
```

```python
#   Creating lists for the DataFrame columns 
# Episode_number
Episode_number = []

# Episode_title
Episode_title = []

# Episode_description
Episode_description = []

# Episode_airdate
Episode_airdate = []

# Episode_review_score
Episode_review_score = []

# Episode_votes
Episode_votes = []
```

```python
#   Create a Microsoft Excel sheet
excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = "HOTD Episodes"
sheet.append(["Episode_number","Episode_title","Episode_description","Episode_Airdate","Episode_review_score","Episode_votes"])
```

```python
#   Get the necessary data for the columns
for episode in episodes:
    #   Episode_number
    number = episode.find("meta", itemprop="episodeNumber")
    print(number["content"] if number else "0")
    e_number = number["content"]

    #   Episode_title
    title = episode.strong.a.text
    print(f"title: {title}")

    #   Episode_description
    description = episode.find("div", attrs={"class": "item_description"}).text.strip()
    print(f"description: {description}")

    #   Episode_Airdate
    # # Use .strip() to remove the spaces at the beginning and end
    # # Use.replace() to remove the "."
    airdate = episode.find("div", attrs={"class": "airdate"}).text.strip().replace(".","")
    print(f"airdate: {airdate}")

    #    Episode_review_score
    review_score = episode.find(
        "div", attrs={"class": "ipl-rating-widget"}).find(
        "div", attrs={"class": "ipl-rating-star small"}).find(
        "span", attrs={"class": "ipl-rating-star__rating"}).text
    print(f"review_score: {review_score}")

    #   Episode_votes
    # # Use .strip() to remove the "()" at the beginning and end
    # # Use.replace() to remove the ","
    votes = episode.find(
        "div", attrs={"class": "ipl-rating-widget"}).find(
        "div", attrs={"class": "ipl-rating-star small"}).find(
        "span", attrs={"class": "ipl-rating-star__total-votes"}).text.strip("()").replace(",","")
    print(f"votes: {votes}")
    print("\n")
    
     #   Store scraped data in Microsoft Excel sheet  
    sheet.append([e_number,title,description,airdate,review_score,votes])
    excel.save("House of the Dragons.xlsx")
    
    Episode_number.append(e_number)
    Episode_title.append(title)
    Episode_description.append(description)
    Episode_airdate.append(airdate)
    Episode_review_score.append(review_score)
    Episode_votes.append(votes)
```

```python
#   Creating a DataFrame for the scraped data
df = pd.DataFrame({
    "Episode_number": Episode_number,
    "Episode_title": Episode_title,
    "Episode_description": Episode_description,
    "Episode_Airdate": Episode_airdate,
    "Episode_review_score": Episode_review_score,
    "Episode_votes": Episode_votes,
})
```
